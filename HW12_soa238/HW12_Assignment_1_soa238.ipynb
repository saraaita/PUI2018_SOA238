# 4. Fourier transform
from the numpy package you will nees only the 
np.fft.rfft() to take the fourier transform

and

np.fft.rfftfreq() to create a meaningful x axis 

Fourier transforms are complex numbers of real or complex imputs. Time series are real numbers so we can use rfft (the Fourier Transform of a **Real imput**, which is itself a complex number)

# to detect periodicity we use the "power spectrum": *square (or absolute value) of the real part of the fourier transform*


`````python
sample_rate = 1 #samples per week
duration = 194 #weeks

N = duration * sample_rate #samples (88)
fig=pl.figure(figsize=(15,5))
ax=fig.add_subplot(111)
ax.set_xticklabels([ "%.2f"%(f) for f in ax.get_xticks()], fontsize=20)
pl.xlabel("frequency (1/week - Hz)", fontsize=20)
pl.ylabel("Power", fontsize=20)
ax.set_title("Power Spectrum", fontsize=20);

for station in range(mta.shape[0]):
    for rtype in range(mta.shape[1]):
        #print(mta[station][rtype])
        #ax.plot(daterange, np.log10(mta[station][rtype]))    
        #ax.plot(np.square((np.fft.rfft(mta[0][0]).real)));
        f = (np.fft.rfft(mta[station][rtype]).real)**2
        ax.plot(np.fft.rfftfreq(N, 1.0/sample_rate)[1:], (f)[1:], 'o', ms=20)
        ax.plot(np.fft.rfftfreq(N, 1.0/sample_rate)[1:], (f)[1:], '-')

`````

Figure 2. MTA Rides Showing Frecuency Peaks

`````python
fig=pl.figure(figsize=(15,5))
ax=fig.add_subplot(111)
pl.xlabel("Frequency [0, ... , 1/(2-i)*sample_rate] Hz", fontsize=20);
pl.legend(fontsize=20)
pl.title ("Real and complect component of the Fourier Transform");
plt.xticks(rotation=70)
for station in range(mta.shape[0]):
    for rtype in range(mta.shape[1]):

        f = np.fft.fft(mta[0][0])
        ax.plot(np.fft.fftfreq(N, 1.0/sample_rate), f, label="fft")
        ax.plot(np.fft.rfftfreq(N, 1.0/sample_rate), np.fft.rfft(mta[0][0]), label="rfft", alpha=0.5)
`````

Figure 3. MTA Rides Showing Frecuency Peaks. 

`````python
fig=pl.figure(figsize=(15,5))
ax=fig.add_subplot(111)
pl.legend(fontsize=20)
ax.set_xlim(0,1.0/15)
ax.set_xticklabels([ "%.2f"%(1/4./f) for f in ax.get_xticks()], fontsize=20)
pl.xlabel("period (weeks)", fontsize=20)
pl.ylabel("Power", fontsize=20)
pl.title("MTA Ride", fontsize=20)

for station in range(mta.shape[0]):
    for rtype in range(mta.shape[1]):
        f = np.abs(np.fft.rfft(mta[0][0]))**2
        ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (f[1:]), 'o', ms=20)
        ax.plot(np.fft.rfftfreq(N, 1.0)[1:], (f[1:]), '-')
`````

`````python
## Aggregate per station
mta_station = np.nanmean(mta,axis=1)
mta_station[0].shape
`````

`````python
fmax_annual=np.zeros(mta_station.shape[0])
top4stations_annual=[]
for station in range(mta_station.shape[0]):
    f = np.abs(np.fft.rfft(mta_station[station])**2)
    fmax_annual[station]=np.nanmax(f[46:58]) 
    #print(sorted(fmax)[-4])
for station in range(mta_station.shape[0]):
    if fmax_annual[station]>=sorted(fmax_annual)[-4]:
        top4stations_annual.append(station)    
`````
